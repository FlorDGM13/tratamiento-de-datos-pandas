##IDENTIFICACION Y EXCLUYENDO OUTLIERS

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize =(14, 6))
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep=';')
datos.head(10)
```

```python
datos.boxplot(['Valor'])
```

```python
datos[datos['Valor'] >=500000]
```

```python
valor = datos['Valor']
```

```python
Q1= valor.quantile(.25)
Q3= valor.quantile(.75)
IIQ= Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 15 * IIQ
```

```python
seleccion = ((valor >= limite_inferior) & (valor <= limite_superior))
```

```python
datos_new = datos[seleccion]
```

```python
datos_new.boxplot(['Valor'])
```

```python
#histograma sirve para mostrar la distribucion de frecuencia de nuestrs datos
datos.hist(['Valor'])
```

```python
datos_new.hist(['Valor'])
```

```python
datos.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = datos.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
Q1= grupo_tipo.quantile(.25)
Q3= grupo_tipo.quantile(.75)
IIQ= Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 15 * IIQ
```

```python
Q1
```

```python
limite_inferior
```

```python
  limite_superior
```

```python
  datos_new = pd.DataFrame()
  for tipo in grupo_tipo.groups.keys():
    eh_tipo = datos['Tipo'] == tipo
    eh_dentro_limite = (datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior)
    seleccion = eh_tipo & eh_dentro_limite
    datos_seleccion = datos[seleccion]
    datos_new = pd.concat([datos_new, datos_seleccion])
```
